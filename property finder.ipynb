```python
import pandas as pd
import numpy as np
import os
```

```python
projects = pd.read_csv('data/project.csv')
addresses = pd.read_csv('data/ProjectAddress.csv')
configurations =pd.read_csv('data/ProjectConfiguration.csv')
variants = pd.read_csv('data/projectconfigurationvariant.csv')
```

```python
print(projects.shape)
projects.head()
```

```python
print(addresses.shape)
addresses.head()
```

```python
projects.isnull().sum()
```

```python
projects =projects.merge(addresses,left_on='id',right_on='projectId')
print(projects.shape)
projects.head()
```

```python
projects.drop('id_x',axis=1,inplace=True)
projects.rename(columns={'id_y':'id'},inplace=True)
projects.info()
```

```python
project_new = projects[['id','projectId','projectType','projectName','status','landmark','fullAddress','pincode']]
```

```python
project_new.head(2)
```

```python
import re 

junk_exp = r'^[a-zA-Z]+$'

project_new['landmark'] = project_new['landmark'].replace(junk_exp,np.nan,regex=True)
#project_new
```

```python
project_new.dropna(subset=['landmark'],inplace=True)
project_new.shape 
```

```python
project_new.head(2)
```

```python
print(configurations.shape)
configurations.head()
```

```python
print(variants.shape)
variants.head()
```

```python
configurations =configurations.merge(variants,left_on='id',right_on='configurationId')
```

```python
print(configurations.shape)
configurations.head(2)
```

```python
config_new = configurations[['projectId','type','price']]
```

```python
print(config_new.shape)
config_new.head(2)
```

```python
project_merge = project_new.merge(config_new,on ='projectId')
print(project_merge.shape)
project_merge.head(2)
```

```python
def map_city(pincode):
    pincode_str = str(pincode)
    if pincode_str.startswith('411') or pincode_str.startswith('412'):
        return 'Pune'
    elif pincode_str.startswith('400'):
        return 'Mumbai'
    else:
        return 'Other'
```

```python
project_merge['city'] = project_merge['pincode'].apply(map_city)
```

```python
print(project_merge.shape)
project_merge.head(2)
```

```python
project_merge['search_text'] = project_merge['projectName'] + ' ' + project_merge['status'] + ' ' + project_merge['landmark'] + ' ' + project_merge['fullAddress'] + ' ' + project_merge['city'] + ' ' + project_merge['type'].astype(str) + ' ' + project_merge['price'].astype(str)
```

```python
project_final = project_merge[['projectName','status','landmark','fullAddress','city','type','price','search_text']]
print(project_final.shape)
project_final.head(10)
```

```python
project_final.to_csv('property_clean.csv',index=False)
```
